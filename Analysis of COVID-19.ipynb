```python
### ANALYSIS OF CORONAVIRUS-19 Wuhan Genome
#Loading our file
from Bio import SeqIO
```

```python
# Doing Analysis
#here equating above info to covid_record
covid_record = SeqIO.read('COVID19 genome Wuhan.fasta','fasta') 
```

```python
#assignging the covid dna sequence nucleotides
covid_dna = covid_record.seq
```

```python
covid_dna
```

```python
#length of the sequence
len(covid_dna)
```

```python
#Protein synthesis, dna to mrna then to proteins
covid_mRNA = covid_dna.transcribe()
```

```python
covid_mRNA
```

```python
#translating the mRNA to protein
covid_protein = covid_mRNA.translate()
```

```python
covid_protein
```

```python
#eseentially its DNA length / 3 cos bcome polypeptides liao
len(covid_protein)
```

```python
# Find all amino acids using split at stop codons
covid_aa = covid_protein.split('*')
```

```python
covid_aa
```

```python
#making it easier to read
covid_aaclean = [str(i) for i in covid_aa]
```

```python
#using pandas which is analysis tool
import pandas as pd #changing the name to shorter pd
```

```python
#creates a nice table showing all the amino acids
df = pd.DataFrame({'Amino Acids':covid_aaclean})
```

```python
df
```

```python
df['length'] = df['Amino Acids'].str.len()
```

```python
df
```

```python
#largest amino acids
#df.nlargest(n rows, "criteria")
df.nlargest(10,"length")
```
